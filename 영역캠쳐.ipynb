```python
import pyautogui
import keyboard
import time
from datetime import datetime
from PIL import Image

# 캡처한 화면을 저장할 디렉토리 설정
save_path = r"C:\\Users\\john5\\Pictures\\Screenshots\\"

# 좌표와 크기로 영역 설정 함수 (듀얼 모니터 지원)
def get_region():
    print("Enter the left, top, width, and height for the capture region:")
    left = int(input("Left: "))
    top = int(input("Top: "))
    width = int(input("Width: "))
    height = int(input("Height: "))

    # 음수 값 확인 및 변환
    if width < 0:
        left += width
        width = abs(width)
    if height < 0:
        top += height
        height = abs(height)

    # 화면 해상도 가져오기
    screen_width, screen_height = pyautogui.size()

    # 영역이 화면을 벗어나지 않는지 확인
    if left < -screen_width or top < -screen_height or left + width > screen_width or top + height > screen_height:
        print("Error: The specified region extends outside the screen boundaries.")
        return None
    if width <= 0 or height <= 0:
        print("Error: Width and height must be positive values.")
        return None

    return (left, top, width, height)

# 화면 캡처 함수 (특정 영역)
def capture_screen_region(region, count):
    left, top, width, height = region

    # 현재 시간으로 파일 이름 생성
    timestamp = datetime.now().strftime("%Y-%m-%d_%H-%M-%S")
    file_name = f"{save_path}screenshot_{timestamp}_{count}.png"

    # 특정 영역 캡처
    screenshot = pyautogui.screenshot(region=region)

    # 파일로 저장
    screenshot.save(file_name)
    print(f"Screenshot {count} saved as {file_name}")

# 매크로 실행 함수
def start_macro():
    print("Press 's' to start capturing screenshots, 'q' to quit.")
    region = None
    while True:
        try:
            if keyboard.is_pressed('s'):
                if region is None:
                    region = get_region()
                    if region is None:
                        break
                for i in range(300):
                    capture_screen_region(region, i + 1)
                    pyautogui.click(x=355, y=355)  # 특정 위치 클릭
                    pyautogui.press('pagedown')  # 페이지다운 키 누르기
                    time.sleep(1)  # 딜레이 추가
                break
            if keyboard.is_pressed('q'):
                print("Quitting...")
                break
        except Exception as e:
            print(f"Error: {e}")
            break

# 매크로 시작
if __name__ == "__main__":
    start_macro()
####355,195,1242,843####
```
